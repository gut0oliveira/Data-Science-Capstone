```python
dados = {
    "IDS-2018-1-COMPLETO": df_2018_1,
    "IDS-2018-2-COMPLETO": df_2018_2,
    "IDS-2018-3-COMPLETO": df_2018_3
}

for nome, df in dados.items():
    val_unico = df.nunique()
    colunas_removidas = val_unico[val_unico == 1].index
    dados[nome] = df.drop(columns=colunas_removidas)
    
    print(f'\n{nome} - Colunas Removidas:\n{list(colunas_removidas)}')
```
